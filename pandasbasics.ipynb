to find min and max in each column


````python
print(df['Maths'].min())
print(df['Maths'].max())
max_marks=df['Maths'].max()
````

````python
df.loc[df['Maths'].idxmax()]
````

````python
df.loc[df['Maths'].idxmax(),"Roll Number"] #loc can find the row when the input is a number
````

````python
df[df['Maths']>max_marks-10]
````

````python
print(df.nlargest(3,"Maths")["Roll Number"])
````

````python
#get a single value
top1=df.nlargest(3,"Maths")["Roll Number"].iloc[0]

#get as a ist
top3=df.nlargest(3,"Maths")["Roll Number"].tolist()


print(top1) #str/int
print(top3) #list
````

# slicing and indexing, **iloc**


```
df.iloc[0]          # first row
df.iloc[1:3]        # rows at position 1 and 2
df.iloc[0, 1]       # first row, second column
df.iloc[:, 0]       # all rows, first column
```



````python
#position-based (like Python lists)
df.iloc[1]
df.iloc[4:8]
df.iloc[4:8,:3]
````

# loc â†’ label-based (real-world usage)

````python
df.loc[0]
df.loc[0:5]
df.loc[0:5,"Maths"]
df.loc[df["Maths"]>70]
df.loc[df["Maths"]>70,["Roll Number","Maths"]]
````
