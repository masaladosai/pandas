```python
print(uploaded_file.keys())
filename=next(iter(uploaded_file))
print(filename)
```

```python
df=pd.read_excel(filename)
df.info()
```

```python
#2 What is the maximum score achieved in Maths? (INT)
df['Maths'].max()
```

```python
#3 What is the minimum score achieved in Physics? (INT)
df['Physics'].min()
```

```python
#4 What is the average score achieved in Chemistry?  (2D)
print(df['Chemistry'].mean())
```

```python
#5 What is the average score achieved in Lang_Adj? (2D)
print(df["Language"].mean()*4)
```

```python
#6 Find the total number of MECH department students who got a D grade. (INT)

#make a new data frqme with average of all subjects
df["average"]=(df[["Maths","Physics","Chemistry"]].sum(axis=1)+df["Language"]*4)/4
df[(df["average"]>=50) & (df["average"]<60) & (df["Department"]=="MECH")].shape[0]
```

```python
#Find the total number of CS department students who got C grade. (INT)
df[(df["average"]>=60) & (df['average']<75) & (df["Department"]=="CS")].shape[0]
```
